#### Plotting multiple lines

```python
# Compare the yeilds of Orange Vs Apples

years = range(2000, 2012)
# Synthetic yields (in tons per hectare)
apples = [0.895, 0.912, 0.930, 0.915, 0.940, 0.955, 0.970, 0.960, 0.975, 0.990, 1.005, 1.010]
oranges = [0.780, 0.795, 0.810, 0.820, 0.835, 0.850, 0.865, 0.875, 0.890, 0.905, 0.920, 0.935]

```

### Chart Title and Legend

To differentiate between multiple lines, we can include a legend within the graph using plt.legend function

We can also give a title to entire chart using plt.title function

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');

plt.title("Crop Yields in Kanto")
plt.legend( ['Apples', 'Oranges'] );
# First you plotted chart for apples so the first value in legend list must be apples as it corresponds with the order in which line plotted
```

### Line Markers

Show markers for the data points on each line using marker argument of plt.plot.

Matplotlib supports many diffrenet types of markers like circles, diamond, cross, square etc

```python
plt.plot(years, apples, marker = 'o')
plt.plot(years, oranges, marker = 'x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend( ['Apples', 'Oranges'] );
```

#### Styling lines and markers of line chart

plt.plot() function supports many arguments for styling lines and markers:

* color or c : set the color of line
* linestyle or ls: choose between a solid or dashed line
* linewidth or lw: set the width of a line
* markersize or ms: set the size of markers
* markeredgecolor or mec: set the edge color for markers
* markeredgewidth or mew: set the edge width for markers
* markeredgecolor or mfc: set the fill color for markers
* alpha: opacity of the plot

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(years, apples, marker = 's', c = 'b', ls = '-', lw = 2, ms = 8, mew = 2, mec= 'navy')
plt.plot(years, oranges, marker = 'o', c = 'r', ls = '--', lw = 3, ms = 10, alpha = .5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend( ['Apples', 'Oranges'] );
```

The *fmt* argument provides a shorthand for specifying the line style, marker and line color.

It can be provided as the third argument to plt.plot

   fmt = '[marker][line][color]'

```python
plt.plot(years, apples, 's-b') # square marker,solid line, color is blue
plt.plot(years, oranges, 'o--r') # circular marker,  dashed line, color is red

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend( ['Apples', 'Oranges'] );
```

## Scatter Plot

In a scatter plot, the values of 2 variables are plotted as points on 2-dimensional grid.

You can use the third variable to determine the size or color of the points

The Iris flower dataset provides samples measurements of sepals and petals for 3 species of flowers. The Iris dataset is included with the Seaborn library, and can be loaded as a Pandasdataframe.

```python
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris") # already there is sns
```

```python
flowers_df
```

```python
# Check all the unique species 
flowers_df.species.unique()
```

Let's try to visualize the relationship between sepal length and sepal width. Our first instinct might be to create a line chart using `plt.plot`. However, the output is not very informative as there are too many combinations of the two properties within the dataset, and there doesn't seem to be simple relationship between them.

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
plt.figure(figsize= (3, 3));
```

#### Find relationship betwen sepal width and sepal length if the length increses then width decreases vica vera

#### We can use a scatter plot to visualize how sepal length & sepal width vary using the *scatterplot* function from seaborn (imported as sns)

matplotlib also provide scatter plot chart but its limited

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=flowers_df)
```

There seem no clear relationship between length and width, the points seem to be all over the place, 
#### We can notice 2 or 3 cluster and some outliers and there are three species of flowers not looking at all species but we can examine relationship btw len and width in one species may be we can find

### Adding Hues
#### Adding Hues : Notice how the points in the above plot seem to form distinct clusters with some outliers. We can color the dots using the flower species as a `hue`. We can also make the points larger using the `s` argument.

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=flowers_df, hue= flowers_df.species, s = 100);
# s is size of dots
```

There seem relationship!
In setosa flower there sepal length tend to be low while sepal length tend to be high as compared to other flower
Virginica tends to have very large length of sepal

Adding hues makes the plot more informative. We can immediately tell that flowers of the Setosa species have a smaller sepal length but higher sepal widths, while the opposite holds true for the Virginica species. 

Seaborn is built on the top of matplotlib, we can use functions like plt.figure and plt.title to modify the figure.

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x='sepal_length', y='sepal_width', data=flowers_df, hue= flowers_df.species, s=100);
```

## Histogram

A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the numbers of observations that fall in each bin.

As an example, let's visualize the how the values of sepal width in the flowers dataset are distributed. We can use the plt.hist function to create a histogram.

```python
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df.sepal_width
```

```python
# lets examine sepal width more
flowers_df.describe()

# but its still not clear where do values lie, how many values lie betwen 2.5 to 3.5
# , if we visualize the distribution we can of the values along a specific range
# sepal width takes the value from 2.0 to 4.4 
```

```python
plt.title('Distribution of Sepal width')
plt.hist(flowers_df.sepal_width);
```

#### We can immediately see that values of sepal width fall in the range 2.0 - 4.5, and around 35 values are in the range 2.9 - 3.1, which seems to be the largest bin.

Entire width taken by sepal is divided into 10 parts , 10 intervals by default.There are around 4 measurements of width 2.0 to 2.2. The highest no of measurements taken are 36 for sepal of width from 3.0 to 3.2. Bin is a mathemetical term , we can control it , each individual bin represnts interval of values, we can split the distribution into 5 bins to clear understanding of values.


### Controlling the size and number of bins

We can control the number of bins, or the size of each bin using the `bins` argument.

```python
plt.hist(flowers_df.sepal_width, bins = 5);
```

We can easily determine the that the smallest bin having a range from 4 to 4.5 

We can specify what should be the point at which bin is created pass a list of to bins argument of `hist` function

```python
import numpy as np
# Craete an array that starts with 2, goes upto 5 not exactly 5 , and it takes a step of 0.25 
np.arange(2, 5, 0.25)
```

```python
import numpy as np

# Specify the boundaries of each bin
plt.hist(flowers_df.sepal_width, bins = np.arange(2, 5, 0.25));
```

```python
# bins of unequal size 

plt.hist(flowers_df.sepal_width, bins = [1, 3, 4, 4.5])
```

#### Still the interval of 3.0 to 4.0 having the highest hight which shows maximum width having values betwwen  3 to 4

### Multiple Histograms

Similar to line charts, we can draw multiple histograms in a single chart. We can reduce the opacity of each histogram, so the the bars of one histogram don't hide the bars for others.

Let's draw separate histograms for each species of flowers.

Set lower opacity for each of the histogram to reduce overlapping among bins

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));

plt.legend(['Setosa', 'Versicolor'])
```

Sepal width of setosa seem to be at higher level means mostly higher than 3.0, versicolor sepal width tend to be at lower side on range 2 to 3

We can also stack multiple histograms on top of one another.

```python
plt.title('Distribution of Sepal Width')
# We can pass multiple series object as alist , specify bins, and set stacked as True
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

## Bar Chart

Bar charts are quite similar to line charts i.e. they show a sequence of values, however a bar is shown for each value, rather than points connected by lines. We can use the `plt.bar` function to draw a bar chart.

```python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]

# line chart
plt.plot(years, apples);
```

Line chart is great to study the the trend but not for comparing individual value

lets plot bar chart

```python
plt.bar(years, apples)
```

It is become very easy to compare the yeil of apples each year

We can draw a line and bar chart together

```python
plt.bar(years, oranges)
plt.plot(years, oranges, 'o--r')
plt.title('Yield of Oranges');

# We can see the trend and compare the values
```

Like histograms, bars can also be stacked on top of one another. We use the `bottom` argument to `plt.bar` to achieve this.

```python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
```

### Bar Plots with Averages

Let's look at another sample dataset included with Seaborn, called "tips". The dataset contains information about the sex, time of day, total bill and tip amount for customers visiting a restaurant over a week.

```python
tips_df = sns.load_dataset("tips");
```

```python
tips_df # data about customers visiting restraunt week days
```

### How does the evrage bill look line in every week day

```python
bill_avg_df = tips_df.groupby('day')[['total_bill']].mean()
bill_avg_df
```

```python
# the above df has day as an index let us plot this on  x asic and mean values on y axis

plt.bar(bill_avg_df.index, bill_avg_df.total_bill);
```

We might want to draw a bar chart to visualize how the average bill amount varies across different days of the week. One way to do this would be to compute the day-wise averages and then use `plt.bar` (try it as an exercise).

However, since this is a very common use case, the Seaborn library provides a `barplot` function which can automatically compute averages.

```python
sns.barplot(x='day', y='total_bill', data=tips_df)
plt.show()
```

That line on top of bar tells variation of values, its called confidence interval(whisker)

t shows how certain we are about the average (mean) value shown by each bar.

By default, Seaborn shows a 95% confidence interval, meaning:

We are 95% confident that the true average value lies within this range.

The lines cutting each bar represent the amount of variation in the values. For instance, it seems like the variation in the total bill was quite high on Fridays, and lower on Saturday.

We can also specify a `hue` argument to compare bar plots side-by-side based on a third feature e.g. sex.

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips_df)
```

You can make bar horizontal by switching the axis

```python
sns.barplot(x = 'total_bill', y = 'day', hue = 'sex', data = tips_df);
```

### Plotting multiple charts in a grid (supported by matplotlib & seaborn)
#### Using `plt.subplots()`, which returns a set of axes that can be used for plotting.

plt.subplots(no of rows, no of columns)
fig, axes are returned by this

```python
fig, axes = plt.subplots(2, 3, figsize = (12, 8))

axes[0, 0].plot(years, oranges, 'o--r')# circular marker , dashed line with red color
axes[0, 0].set_title('Yeilds of oranges')

# axes[0, 2].plot(years, apples, 's-g')# circular marker , dashed line with red color
# axes[0, 2].set_title('Yeilds of apples')

axes[1, 1].plot(years, oranges, 'o--b')
axes[1, 1].plot(years, apples, 'x-r')
axes[1, 1].set_title('Yeilds of apples VS oranges')
axes[1, 1].legend(['oranges', 'apples'])
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('Yield (tons per hectare)')

# Draw scatterplaot using with its ax 
# scatterplot provide ax argument to specify the axes on which to draw  
sns.scatterplot(
    data=flowers_df,
    x='sepal_length', # visualize relationship between two variables
    y='sepal_width',
    hue='species',
    s=100,
    ax=axes[1, 2]
)

axes[1, 2].set_title('Sepal length vs width')

axes[0, 1].hist( # distribution of a value of a single variables
    [setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width ] ,  # pass array/Series directly
    bins=np.arange(2, 5, 0.25),
    stacked=True
)
axes[0, 1].legend(['Setosa', 'Versicolor', 'Verginica']);


# bar plot
axes[0, 2].set_title('Restaurant bills')

sns.barplot(
    x='day', # avg total bill on each weekday
    y='total_bill',
    hue='sex',
    data=tips_df,
    ax=axes[0, 2]
)

axes[1, 0].set_title('Flight traffic')

sns.heatmap(
    flights_df,
    cmap='Blues', # flight traffic on a perticular airport at a time
    ax=axes[1, 0],
    annot=True,
    fmt='d'
)

# Plot an image using the axes
# axes[1, 2].set_title('Image Plotting on an axes');
# axes[1, 2].imshow(img)
# axes[1, 2].grid(False)
# axes[1, 2].set_xticks([])
# axes[1, 2].set_yticks([])


plt.tight_layout(pad = 2) # this gives space between these boxes
#  To start filling these graphes use results of subplots

```

## Branching with `if`, `else` and `elif`
Branching a the ability to make decision and execute a different set of statements based on whether one or more conditions are true.

### The `if` statement
In python, branching is done using if statement, which has the following syntax:
   if `condition`
      statement1
      statement2

the condition can either be a variable or an expression. If the condition evaluates to True,
then the statement within the if block are executed. `statement2` is also associated with the if statement.

The technique of structuring code by adding spaces is called `indentation`.
All the statements that rae indentated belong to if block.

Python relies heavily on indentation(white space before a statement) to define structure in code. This makes Python code easy to read and understand.placimg the cursor at the start of the line and pressing the `Tab` key once to add 4 spaces

```python
a_number = 34
```

```python
if a_number % 2 == 0:
    print("We're inside an if block")
    print('The given number {} is even.'.format(a_number))
```

Note that we are using the modulus operator `%` to find the remainder from the division of `a_number` by `2`, and then we are using the comparision operator `==` check if the reminder is `0`, indicating that the number is divisible by 2 i.e it is even.

The number 34 is indded divisible by 2 the expression a_number % 2 == 0 evaluated to `True`, so the print statements inside under the if statement is executed, If the a_number was 35, then condition would have evaluated to False and statements would not have execeuted.

```python
another_number = 33
```

```python
if another_number % 2 == 0:
    print('The given number {} is even.'.format(another_number))
```

### The `elif` statement

Python also provides an `elif` statement to chian a series of conditional blocks. the conditions are evaluated one by one. For the first condition that evaluates to `True`, the statements in the respective block are executed, and the remaining conditions are not evaluated. So in the chain of `if`, `elif`, `elif`.., exactly one conditional block that first evaluates to true is executed.

As the condition keep failing we check next condition until we find true condition

```python
today = 'Wednesday'
```

```python
if today == 'Sunday':
    print("Today is the day of sun.")
elif today == 'Monday':
    print("Today is the day of the monn.")
elif today == 'Tuesday':
    print("Today is the day of Tyre, the god of war")
elif today == 'Wednesday':
    print("Today is the day odin, the supreme diety.")
elif today == 'Thursday':
    print("Today is the day of thor, the god of thunder")
elif today == 'Friday':
    print("Today is the goddess of beauty.")
```

The first 3 conditions are evaluated to True so none of the first 3 messages are printed. The fourth condition is evaluated to True, so the corresponding message is printed. The remaining conditions are skipped.

### Using `if`, `elif` and `else` together
You can also include an else statement at the end of chain of if, elif.. statements. This code within the else block are evaluated when none of the conditions hold true.

# Iteration with while loops
Allows us to execute statement or bunch of statements multiple times.
Iterating or looping is the ability to run one or more statements multiple times
There are two ways to do this inPython:
`while` loops and `for` loops

while loop has the following syntax:
  
  while condition:
    statements(s)

The statements in block under `while` are executed repetedly as long as the `condition` is `True`. In most cases, the block of statements makes some changes to a vaiable which causes the condition to evaluate to `False` after a certain number of iteration.

```python
# factorial 
result = 1
i = 1

while i <= 100:
    result = result * i
    i = i + 1

print("The factorial of a number {}".format(result)) # very quickly we calculated factorial of 100
```

```python
%%time 
 # this will tell us how long it takes to execute this cell

result = 1
i = 1

while i <= 1000:
    result *= i
    i += i

print(result)
```

Using two while loop to create an interesting pattern

```python
line = '*'
max_length = 10

while len(line) <= max_length:
    print(line)
    line += '*'

while len(line) > 0:
    print(line)
    line = line[:-1] #  [: -1] start is beginning of string and end is one character before the last
    # it removes last character from line each time
```

### Iteration with `for` loops

A for loop is used fior iterating or looping over sequence i.e. lists, tuples, dictionaries, strings and ranges. For loops have the following syntax:
   for value in sequence:
      statements(s)

The statements within the loop are executed once for each element in `sequence`. 

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thurday', 'Friday']

for day in days:
    print(day)
    
```

```python
# Looping over a string
for char in 'Monday':
    print(char)
```

```python
# Looping over a tuple
for fruit in ['Apple', 'Banana', 'Guava']:
    print("Here's a fruit: ", fruit)
```

```python
# Looping over a dictionary
person = {
    'name': 'John Doe',
    'sex': 'Male',
    'age': 32,
    'married': True
}

for key in person:
    print('key: ', key, ",", "value: ", person[key])
```

While using dictionary with for loop , the iteration happens over keys within the dictionary. The key can be used within the loop to access the value. You can directly iterate over the values using the `.values` method of the dictionary, or over key-value pair using the `.items` method

```python
for value in person.values():
    print(value)
```

```python
for key_value_pair in person.items():
    print(key_value_pair) # a tuple
```

#### Iterating using `range` and `enumerate`
The `range` function is used to create a sequence of numbers that can be iterated over using a `for` loop. It can be used in 3 ways:
   * **range(n)** - Creates a sequence of numbers from 0 to n - 1.
   * **range(a, b)** - Creates a sequence of numbers from a to b - 1.
   * **range(a, b, step)** - Creates a sequence of numbers from a to b - 1 with increment of `step`

```python
for i in range(7):
    print(i)
```

```python
for i in range(3, 10):
    print(i)
```

```python
for i in range(3, 14, 4):
    print(i)
```

Ranges are used for iterating over lists, when you need to track the index of elements while iterating

```python
a_list = ['Monday', 'Tuesday', 'Wednesday', 'Thurday', 'Friday']

for i in range(len(a_list)):
    print("The value at position {} is {}".format(i, a_list[i]))
```

Another way to achieve the same result as above is using the **`enumerate`** function with a_list as an input, which returns a tuple containing the index and the corresponding element.

```python
for i, val in enumerate(a_list):
    print('The value at position {} is {}.'.format(i, val))
```

### `break`,`continue` and `pass` statements
`for` loop similar to `while` loops also support the **break** statement for breaking out of the loop and **continue** for skipping ahead to the next iteration.

```python
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thurday', 'Friday']

for day in weekdays:
    print("Today is {}".format(day))
    if (day == 'Wednesday'):
        print("I dont't work beyond Wednesday!")
        break
```

```python
for day in weekdays:
    if (day == 'Wednesday'):
        print("I dont't work on Wednesday!")
        continue
    print("Today is {}".format(day))
```

Similar to `if` statement, `for` loops cannot be empty, so you can use a **pass** statement in case you do not want to execute any statement inside the loop

```python
for day in weekdays:
    pass
```

#### Documenting functions using Docstring
We can add some documentation within our function using *docstring*. A *docstring* is simply a string that appears as the first statement within the function body, and is used by the *help* function. A good *docstring* describes what the function does, and provides some explaination about documents.

```python
def loan_emi(amount, duration, rate, down_payment = 0):
    """ Calculates the Equal Monthly Installment for a loan.
    
    Arguments:
       amount - Total amount to be spent(loan + down payment)
       duration - Duration of the loan
       rate - rate of interest (monthly)
       down_payment - optional initial payment"""
    loan_amount = amount - down_payment
    try:
         emi = loan_amount * rate * ((1 + rate)**duration) / (((1 + rate)**duration - 1))
    except ZeroDivisionError:
         emi = loan_amount / duration      
    return math.ceil(emi)
```
